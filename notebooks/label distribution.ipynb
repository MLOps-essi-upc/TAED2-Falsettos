# Data label distribution
Code designed to see the data distribution from dataset. First we import libraries.

```python
import os

import datasets

import numpy as np
import matplotlib.pyplot as plt 
```

We get the paths to load the dataset correctly. 

```python
path = os.getcwd()
ROOT_DIR =  os.path.dirname(path)
RAW_DATA_DIR = os.path.join(ROOT_DIR, 'src', 'data', 'raw')
PROCESSED_DATA_DIR = os.path.join(ROOT_DIR, 'src', 'data', 'processed')
```

We load the dataset and we get the labels.

```python
audio_dataset = datasets.load_from_disk(RAW_DATA_DIR)
labels = audio_dataset['train']['label'] + audio_dataset['validation']['label'] + audio_dataset['test']['label'] # list of all splits
print("Number of samples:", len(labels))
print("Number of labels:", len(set(labels)))
```

We obtain the label distribution.

```python
label_count = {i:labels.count(i)/len(labels) for i in labels}
```

And finally we plot the results. 

```python
labels = list(label_count.keys())
values = list(label_count.values())
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(labels, values, color ='maroon', 
        width = 0.4)

plt.xlabel("Label id")
plt.ylabel("No. of data")
plt.title("Label data distribution")
plt.show()
```

We can see that the dataset is unbalanced. We have to take this into account when we are training the model and choosing the loss function and validation metrics.
